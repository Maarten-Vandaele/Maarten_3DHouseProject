```python
lat = dms2dec(latDMS)
```

## transformer to lambert

```python
transformer = pyproj.Transformer.from_crs('epsg:4326', 'epsg:31370')
lambtuple = transformer.transform(lat,lon)
```

```python
# print(lambtuple[1])
```

```python
# quarterbox = (int(box)//2)
# print(quarterbox)
```

```python
left = lambtuple[0] - 20#(int(box)//2)    ##bounding box choosing option abandoned
right = lambtuple[0] + 20#(int(box)//2)
bottom = lambtuple[1] - 20#(int(box)//2)
top = lambtuple[1] + 20#(int(box)//2)

with rasterio.open("./assets/DHMVIIDSMRAS1m_k13.tif") as src:
    rst = src.read(1, window=from_bounds(left, bottom, right, top, src.transform))
    
# rasterio.plot.show(rst, cmap='bwr')
```

```python
df_window = pd.DataFrame(rst)
# type(df_window)
# df_window.iloc[:-1,:-1] = df_window.iloc[:-1,:-1]*.7
# df_window.head()
```
